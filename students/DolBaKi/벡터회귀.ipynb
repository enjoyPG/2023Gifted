```python
def update():
  global cost
  cost=0
  for i in range(5):
    global w1
    global w2
    global bias
    w1=w1-alpha*(2*x[i][0]*x[i][0]*w1 + 2*w2*x[i][0]*x[i][1] + 2*bias*x[i][0] - 2*x[i][0]*y[i])
    w2=w2-alpha*(2*x[i][1]*x[i][1]*w2 + 2*w1*x[i][0]*x[i][1] + 2*bias*x[i][1] - 2*x[i][1]*y[i])
    bias=bias-alpha*(2*bias + 2*w2*x[i][1] + 2*w1*x[i][0] - 2*y[i])
  for i in range(5):
    cost+=(w1*x[i][0] + w2*x[i][1] + bias - y[i]) * (w1*x[i][0] + w2*x[i][1] + bias - y[i])

import numpy as np
import matplotlib.pyplot as plt
x=np.array([[80,7],[80,5],[130,10],[70,9],[70,10]])
y=np.array([65,51,50,82,78])
global alpha
alpha=1e-5 * 5
w1=0
w2=0
bias=0

epoch= int(input())
for i in range(epoch):
  cost=0
  update()
  cost=cost/5
  if i % 100000 == 0:
      print(cost,i)
      print(w1,w2,bias,i)

xx=np.array(range(0,140))
plt.plot(xx,w1*xx+bias)
for i in range(5):
  plt.scatter(x[i][0],y[i])
plt.show()
xxx=np.array(range(0,10))
plt.plot(xxx,w2*xxx+bias)
for i in range(5):
  plt.scatter(x[i][1],y[i])
plt.show()
```
